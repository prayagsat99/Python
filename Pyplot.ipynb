```python
x_data=[1,2,3,4,5]
y_data=[1,4,9,16,25]
x1_data=[2,4,6,8,20]
y1_data=[3,6,9,12,15]
plt.plot(x_data,y_data,'o--g')
plt.plot(x1_data,y1_data,'v--r')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis([0,20,0,25])
#plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.show()
```

```python
import numpy as np
x_data=np.linspace(0,10,100)
y1=x_data**2
y2=x_data**3
y3=x_data**4
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("My plot")
plt.plot(x_data,y1)
plt.plot(x_data,y2)
plt.plot(x_data,y3)
plt.show()
```

```python
import numpy as np
x_data=np.linspace(0,10,100)
y1=x_data**2
y2=x_data**3
y3=x_data**4
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("My plot")
plt.plot(x_data,y1, label="square")
plt.plot(x_data,y2, label="cube")
plt.plot(x_data,y3, label="quad")
plt.legend(loc="best")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
y1=np.random.randn(1000)
y2=np.random.randn(1000)
plt.scatter(y1,y2,marker="v",c="r")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
x1=np.random.randn(1000)
x2=np.random.randn(1000)
y1=np.random.randn(1000)
y2=np.random.randn(1000)
plt.scatter(x1,y1,marker="v",c="r")
plt.scatter(x2,y2,marker="o",c="g")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("My Scatter Plot")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
countries=['USA','Brazil','Russia','India','UK']
no_of_cases=[2779953,1453369,654405,605220,313483]
colors=['r','g','b','r','y']
plt.bar(countries,no_of_cases,color=colors)
plt.xlabel("countries")
plt.ylabel("number of cases")
plt.title("Corona cases worldwide")
plt.show()
```
